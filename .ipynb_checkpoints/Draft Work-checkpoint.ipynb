```python
# Apply Canny edge detection (optional)
edges = cv2.Canny(gray_image, threshold1=200, threshold2=500)
plt.imshow(edges)
```

```python
# Find contours in the image
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
plt.imshow(_)
```

```python

# Filter contours based on area (adjust the area threshold as needed)
min_contour_area = 1000
filtered_contours = [contour for contour in contours if cv2.contourArea(contour) > min_contour_area]

```

```python
# Get the coordinates of the bounding box of the first filtered contour (assuming it's the rectangle)
if filtered_contours:
    x, y, w, h = cv2.boundingRect(filtered_contours[0])
    x1, y1, x2, y2 = x, y, x + w, y + h
    print("Top-left coordinates (x1, y1):", x1, y1)
    print("Bottom-right coordinates (x2, y2):", x2, y2)
else:
    print("No suitable rectangle found.")
```
